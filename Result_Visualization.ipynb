CO2 Emission and File Size Reduction:

```python
import matplotlib.pyplot as plt

# Data for CO2 Emission and File Size Reduction
categories = ['Long Code', 'Short Code']
co2_emissions = [1.943, 1.593]  # CO2 emissions in grams
file_sizes = [6.11, 5.01]  # File sizes in MB

# Graph 1: CO2 Emission and File Size Reduction
plt.figure(figsize=(10, 6))

# Plotting CO2 emissions
plt.subplot(1, 2, 1)
plt.bar(categories, co2_emissions, color=['blue', 'orange'])
plt.title('CO2 Emission (g)')
plt.ylabel('CO2 Emission (g)')

# Adding numbers above the bars
for i, emission in enumerate(co2_emissions):
    plt.text(i, emission + 0.05, str(emission), ha='center', va='center_baseline')

# Plotting file sizes
plt.subplot(1, 2, 2)
plt.bar(categories, file_sizes, color=['blue', 'orange'])
plt.title('File Size (MB)')
plt.ylabel('File Size (MB)')

# Adding numbers above the bars
for i, size in enumerate(file_sizes):
    plt.text(i, size + 0.05, str(size), ha='center', va='bottom')

plt.show()
```

Code Processing Speed:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Code Processing Speed
libraries = ['Matplotlib', 'Plotly', 'Bokeh']
long_processing_speed = [0.490, 0.341, 0.428]  # Processing speed in seconds for long code
short_processing_speed = [0.260, 0.093, 0.087]  # Processing speed in seconds for short code

# Adjust the x positions for short code bars
x_long = np.arange(len(libraries))
x_short = x_long + 0.35  # Move the short code bars to the right of the long code bars

# Graph: Code Processing Speed
plt.figure(figsize=(8, 5))

# Plotting long code processing speed
plt.bar(x_long, long_processing_speed, label='Long Code', color='blue', width=0.35)

# Plotting short code processing speed
plt.bar(x_short, short_processing_speed, label='Short Code', color='orange', width=0.35, alpha=0.6)

# Adding numbers above the bars
for i, speed in enumerate(long_processing_speed):
    plt.text(x_long[i], speed + 0.005, f'{speed:.3f}', ha='center', va='bottom')

for i, speed in enumerate(short_processing_speed):
    plt.text(x_short[i], speed + 0.005, f'{speed:.3f}', ha='center', va='bottom')

plt.xlabel('Library')
plt.ylabel('Processing Speed (s)')
plt.title('Code Processing Speed (seconds)')
plt.xticks(x_long + 0.17, libraries)
plt.legend()
plt.tight_layout()
plt.show()
```

Memory and CPU Utilization:

```python
import matplotlib.pyplot as plt

# Data for Memory and CPU Utilization
libraries = ['Matplotlib', 'Plotly', 'Bokeh']
code_lengths = ['Long Code', 'Short Code']

memory_usage = {
    'Matplotlib': [110788, 110748],
    'Plotly': [111136, 110544],
    'Bokeh': [167908, 167908]
}

# Calculate average memory usage for long and short code
avg_memory_long = sum(memory_usage[library][0] for library in libraries) / len(libraries)
avg_memory_short = sum(memory_usage[library][1] for library in libraries) / len(libraries)

# Detailed memory usage for each library
long_memory_detailed = [memory_usage[library][0] for library in libraries]
short_memory_detailed = [memory_usage[library][1] for library in libraries]

# Graph: Memory Usage Comparison (Line Chart)
plt.figure(figsize=(10, 6))

# Plotting the detailed memory usage for each library
plt.plot(libraries, long_memory_detailed, marker='o', label='Long Code', color='blue')
plt.plot(libraries, short_memory_detailed, marker='o', label='Short Code', color='orange')

# Add average memory usage lines
plt.axhline(y=avg_memory_long, linestyle='--', color='blue', label=f'Avg Long Code: {avg_memory_long:.0f}')
plt.axhline(y=avg_memory_short, linestyle='--', color='orange', label=f'Avg Short Code: {avg_memory_short:.0f}')

# Add labels and title
plt.xlabel('Library')
plt.ylabel('Memory Usage')
plt.title('Memory Usage Comparison')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for Memory and CPU Utilization
libraries = ['Matplotlib', 'Plotly', 'Bokeh']
code_lengths = ['Long Code', 'Short Code']

cpu_time = {
    'Matplotlib': [2.485927, 2.258821],
    'Plotly': [2.094543, 3.473306],
    'Bokeh': [4.320924, 4.590133]
}

# Graph: CPU Time Comparison
plt.figure(figsize=(10, 6))

# Plotting CPU time for each library
for i, library in enumerate(libraries):
    plt.bar([x + i * 0.2 for x in range(len(code_lengths))], cpu_time[library], width=0.2, label=library)

# Adding numbers above the bars
for i, library in enumerate(libraries):
    for j, time in enumerate(cpu_time[library]):
        plt.text(j + i * 0.2, time + 0.05, f'{time:.2f}', ha='center', va='bottom')

plt.xlabel('Code Length')
plt.ylabel('CPU Time')
plt.title('CPU Time Comparison')
plt.xticks([0.2, 1.2], code_lengths)
plt.legend()
plt.tight_layout()
plt.show()
```
