# 1. Matplotlib library

## 1.1 Actual Code

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 15]

# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Matplotlib')
plt.show()

import cProfile
import resource

# Define the function or code segment to be profiled
def my_function():
    total = 0
    for i in range(1000000):
        total += i
    return total

# Profile the execution of the function
profiler = cProfile.Profile()
profiler.enable()
result = my_function()
profiler.disable()

# Get resource usage statistics
usage_stats = resource.getrusage(resource.RUSAGE_SELF)

# Print the profiling results
profiler.print_stats()
print("Resource usage (in bytes):")
print(f"CPU time: {usage_stats.ru_utime} user, {usage_stats.ru_stime} system")
print(f"Memory usage: {usage_stats.ru_maxrss} maximum resident set size")
```

# Execution Time Comparison

```python
import time
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 15]

# Matplotlib
start_time = time.time()
plt.bar(categories, values, color='skyblue')
plt.show()
matplotlib_execution_time = time.time() - start_time
print("Matplotlib Execution Time:", matplotlib_execution_time)

# Plotly
start_time = time.time()
fig = go.Figure(go.Bar(x=categories, y=values))
fig.show()
plotly_execution_time = time.time() - start_time
print("Plotly Execution Time:", plotly_execution_time)

# Bokeh
start_time = time.time()
p = figure(x_range=categories, title="Bar Chart using Bokeh")
p.vbar(x=categories, top=values, width=0.9)
show(p)
bokeh_execution_time = time.time() - start_time
print("Bokeh Execution Time:", bokeh_execution_time)
```
